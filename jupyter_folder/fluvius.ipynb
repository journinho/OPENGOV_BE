```python
# Web API Using Requests and JSON
import requests
import json
import pandas as pd
```

```python
df_url = 'https://opendata.fluvius.be/explore/dataset/1_06b-verbruiksgegevens-statistische-sector-met-nace-sector/download/?format=csv&timezone=Europe/Brussels&lang=en&use_labels_for_header=true&csv_separator=%3B'
df = pd.read_csv(df_url,sep = ';')
df.head()
```

```python
Aardgas = df[ df['Markt'] == 'Aardgas' ]
Aardgas.head()
```

```python
Elektriciteit = df[(df['Markt'] == 'Elektriciteit') & (df['Verbruiksjaar'] == 2021)]
Elektriciteit.head()
```

```python
elektriciteitsgebruik_sector = pd.pivot_table(Elektriciteit, index = ['Sector'], values= 'Benaderend Verbruik (kWh)', aggfunc = sum)
elektriciteitsgebruik_sector.head(20)
```
