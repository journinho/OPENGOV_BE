```python
import pandas as pd
```

```python
df_url = 'https://www.ecb.europa.eu/stats/eurofxref/eurofxref-hist.zip'
df = pd.read_csv(df_url, compression='zip', header=0, sep=',', decimal=".")
```

```python
# sort data by date
df = df.sort_values(by=['Date'])
# keep only last 5 years
df = df.tail(5*365)
df
```

```python
# get all column names and save as list
cols = df.columns.tolist()
#remove first and last value from list
cols = cols[1:-1]
cols
```

```python
# for every value in cols, create a new dataframe with date and column name
for col in cols:
    df_temp = df[['Date', col]]
    df_temp.to_csv('fx_hist_EUR_' + col + '.csv', index=False)
```
